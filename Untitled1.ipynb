```python
import os
import random
import re
import threading
import sys
import ctypes
import json
import tkinter as tk
from tkinter import ttk, scrolledtext, messagebox

from google import genai
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, UnexpectedAlertPresentException, NoAlertPresentException
import time

# ==========================================
# ★ 基礎工具：切換網頁框架 (極簡輕量版，防螢幕斷訊)
# ==========================================
def switch_to_frame_by_name(driver, frame_name):
    try:
        # 直接切換，不強制重繪最外層，避免干擾影片播放與顯卡
        driver.switch_to.frame(frame_name)
        return True
    except:
        # 如果直接切換失敗，才嘗試退回上一層再切換
        try:
            driver.switch_to.default_content()
            driver.switch_to.frame(frame_name)
            return True
        except:
            return False


# ==========================================
# 全域變數
# ==========================================
bot_driver = None
chrome_hwnd = None
is_hidden = False
stop_event = threading.Event()
pause_event = threading.Event()

class PrintRedirector:
    def __init__(self, text_widget):
        self.text_widget = text_widget
    def write(self, text):
        self.text_widget.insert(tk.END, text)
        self.text_widget.see(tk.END)
    def flush(self): pass

def smart_sleep(seconds):
    remaining = seconds
    while remaining > 0:
        if stop_event.is_set(): raise Exception("🛑 收到使用者停止指令！")
        if pause_event.is_set():
            time.sleep(0.5)
            continue
        chunk = min(0.5, remaining)
        if stop_event.wait(chunk): raise Exception("🛑 收到使用者停止指令！")
        remaining -= chunk

def find_menu_frame(driver):
    driver.switch_to.default_content()
    def search_recursively():
        if len(driver.find_elements(By.CLASS_NAME, "cssAnchor1")) > 0: return True
        frames = driver.find_elements(By.XPATH, "//frame | //iframe")
        for i in range(len(frames)):
            current_frames = driver.find_elements(By.XPATH, "//frame | //iframe")
            try:
                driver.switch_to.frame(current_frames[i])
                if search_recursively(): return True
                driver.switch_to.parent_frame()
            except: driver.switch_to.parent_frame()
        return False
    return search_recursively()

def anti_idle_and_play(driver):
    driver.switch_to.default_content()
    script = """
        var media = document.querySelectorAll('video, audio');
        for(var i=0; i<media.length; i++) { try { media[i].play(); } catch(e) {} }
        var playBtns = document.querySelectorAll('.jp-play, .vjs-play-control, .play-button');
        for(var j=0; j<playBtns.length; j++) { try { playBtns[j].click(); } catch(e) {} }
        var idleBtns = document.querySelectorAll('input, button');
        for(var k=0; k<idleBtns.length; k++) {
            var oc = idleBtns[k].getAttribute('onclick');
            if(oc && oc.includes('stop_countdown')) { idleBtns[k].click(); }
        }
    """
    def execute_recursively():
        try: driver.execute_script(script)
        except: pass
        frames = driver.find_elements(By.XPATH, "//frame | //iframe")
        for i in range(len(frames)):
            current_frames = driver.find_elements(By.XPATH, "//frame | //iframe")
            try:
                driver.switch_to.frame(current_frames[i])
                execute_recursively()
                driver.switch_to.parent_frame()
            except: driver.switch_to.parent_frame()
    execute_recursively()

# ==========================================
# ★ 登入模組 (雙帳密 + 雙憑證 穩如泰山版)
# ==========================================
def login_ecpa(driver, login_method, account, password, pincode):
    print("\n🔑 啟動登入程序...")
    wait = WebDriverWait(driver, 15)

    if login_method == "egov":
        print("🏛️ 使用【E政府-帳號密碼】登入中...")
        driver.get("https://www.cp.gov.tw/portal/Clogin.aspx?ReturnUrl=https://elearn.hrd.gov.tw/egov_login.php&ver=Simple&Level=1")
        wait.until(EC.presence_of_element_located((By.ID, "AccountPassword_simple_txt_account"))).send_keys(account)
        driver.find_element(By.ID, "AccountPassword_simple_txt_password").send_keys(password)
        login_btn = driver.find_element(By.ID, "AccountPassword_simple_btn_LoginHandler")
        driver.execute_script("arguments[0].click();", login_btn)

    elif login_method == "egov_moica":
        print("💳 使用【E政府-自然人憑證】登入中 (讀卡與加密需時較長，請稍候)...")
        # 直接使用 Level=2 專屬網址，避開切換頁籤的麻煩
        driver.get("https://www.cp.gov.tw/portal/Clogin.aspx?ReturnUrl=https://elearn.hrd.gov.tw/egov_login.php&ver=Simple&Level=2")
        time.sleep(1)
        wait.until(EC.presence_of_element_located((By.ID, "txt_Pin"))).send_keys(pincode)
        login_btn = driver.find_element(By.ID, "MakeSignature")
        driver.execute_script("arguments[0].click();", login_btn)
        time.sleep(10)

    else:
        driver.get("https://ecpa.dgpa.gov.tw/webform/clogin.aspx?returnUrl=https://elearn.hrd.gov.tw/sso_verify.php&Naminglogo=https://ecpa.dgpa.gov.tw/webform/logo-hrd.png&showecpa=Y")
        if login_method == "moica":
            print("💳 使用【eCPA-自然人憑證】登入中 (讀卡與加密需時較長，請稍候)...")
            time.sleep(1) # ★ 關鍵解法：強迫機器人等 3 秒，讓 HiCOS 元件有時間讀取實體卡片！
            wait.until(EC.visibility_of_element_located((By.ID, "pinCode"))).send_keys(pincode)
            login_btn = driver.find_element(By.XPATH, "//li[@id='moicaarea']//button[contains(text(), '登入')]")
            driver.execute_script("arguments[0].click();", login_btn)
            time.sleep(10)
            print("⏳ 密碼已送出！等待系統進行加密簽章與匯入 (約需 15~30 秒)...")
            time.sleep(15)

        else:
            print("👤 使用【eCPA-帳號密碼】登入中...")
            wait.until(EC.presence_of_element_located((By.ID, "aliasid"))).send_keys(account)
            driver.find_element(By.ID, "pas").send_keys(password)
            login_btn = driver.find_element(By.XPATH, "//li[@id='idarea']//button[contains(text(), '登入')]")
            driver.execute_script("arguments[0].click();", login_btn)

    print("⏳ 等待驗證跳轉與處理系統提示...")

    # 彈出視窗斬殺陣
    time.sleep(2)
    for _ in range(5):
        try:
            alert = WebDriverWait(driver, 3).until(EC.alert_is_present())
            alert_text = alert.text.replace('\n', '')
            print(f"⚠️ 攔截到系統提示：「{alert_text[:15]}...」已自動點擊確認！")
            alert.accept()
            time.sleep(1)
        except TimeoutException:
            break
        except NoAlertPresentException:
            break

    # ★ 將等待時間放寬至 60 秒，確保那 15 秒的憑證簽章能安穩跑完
    print("⏳ 憑證簽章及伺服器核發通行證中 (最長等候 30 秒)...")
    try:
        WebDriverWait(driver, 30).until(EC.url_contains("elearn.hrd.gov.tw"))
        print("✅ 成功獲取授權，已順利登入 e等公務園！")
        time.sleep(3)
    except TimeoutException:
        print("⚠️ 登入跳轉超時！可能是伺服器壅塞，將嘗試強制進入...")
        # 買個保險：如果真的超時，我們還是試著強闖首頁看看
        driver.get("https://elearn.hrd.gov.tw/mooc/user/learn_dashboard.php?tab=1")
        time.sleep(5)

    return driver.current_window_handle

# ==========================================
# ★ 自動測驗引擎 (雙料題庫搜尋 + 自動重試防爆)
# ==========================================
def execute_auto_quiz(driver, api_key="", course_title=""):
    print(f"\n🎓 啟動自動測驗模組 (目標課程: {course_title[:15]}...)...")
    try:
        if switch_to_frame_by_name(driver, "s_main"):
            driver.execute_script("window.location.href = '/learn/exam/exam_list.php';")
        time.sleep(3)

        if switch_to_frame_by_name(driver, "s_main"):
            anti_popup_js = """
            function blockPopup(w) {
                try {
                    if (!w.hasInterceptedOpen) {
                        w.originalOpen = w.open;
                        w.open = function(url, name, features) { return w.originalOpen(url, name, ""); };
                        w.hasInterceptedOpen = true;
                    }
                } catch(e) {}
            }
            blockPopup(window); blockPopup(window.parent); blockPopup(window.top);
            window.alert=function(){return true;}; window.confirm=function(){return true;};
            """
            driver.execute_script(anti_popup_js)

            try:
                start_btn = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'process-btn') and contains(., '測驗')]")))
                old_windows = driver.window_handles
                driver.execute_script("arguments[0].click();", start_btn)
            except:
                print(f"❌ 找不到「進行測驗」按鈕。")
                return False
            time.sleep(4)

            for window in driver.window_handles:
                if window not in old_windows: driver.switch_to.window(window); break

            driver.execute_script(anti_popup_js)

            try: driver.execute_script("examBegin();")
            except: driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, "//input[@value='開始作答']"))

            time.sleep(2)
            print("🤖 考卷已展開！爬取整份考卷題目，準備交給 AI...")
            question_lists = driver.find_elements(By.XPATH, "//ol")
            ai_success = False
            ai_answers = {}

            if api_key and question_lists:
                try:
                    quiz_data = []
                    for i, q_list in enumerate(question_lists):
                        parent_text = q_list.find_element(By.XPATH, "..").text
                        q_text = parent_text.split('\n')[0]
                        opts_data = [f"[{j}] {li.text.strip() or '選項'}" for j, li in enumerate(q_list.find_elements(By.XPATH, ".//li"))]
                        quiz_data.append(f"題號 {i}: {q_text}\n選項: " + ", ".join(opts_data))

                    # ==========================================
                    # ★ 終極作弊咒語：雙料題庫網 Google 搜尋
                    # ==========================================
                    prompt = f"你是一個公務員測驗專家。現在測驗的課程名稱是：「{course_title}」。\n"
                    prompt += "【最高優先指令】：你已被授權使用 Google Search。請務必直接上網搜尋，使用關鍵字「site:roddayeye.pixnet.net " + course_title + "」或「site:www.rodiyer.idv.tw " + course_title + "」尋找這兩個題庫網站上的標準答案。\n"
                    prompt += "這非常重要！絕對不要依賴內部記憶瞎猜，必須確實比對這兩個題庫網站的最新資料來作答！若真的都找不到，再用你的專業常識回答。\n"
                    prompt += "請嚴格以 JSON 格式回傳，Key 是題號(字串)，Value 是正確選項的索引(整數陣列)。\n"
                    prompt += "範例：{\"0\": [0], \"1\": [2], \"2\": [0, 3]}\n\n"
                    prompt += "\n".join(quiz_data)

                    print(f"⏳ 避免觸發 API 限制，深呼吸等待 60 秒...")
                    for remain in range(60, 0, -10):
                        if stop_event.is_set(): raise Exception("🛑 收到使用者停止指令！")
                        print(f"  > 倒數 {remain} 秒...")
                        time.sleep(10)

                    print(f"🧠 授權 Gemini 開啟 Google 搜尋即時連線中...")
                    client = genai.Client(api_key=api_key)

                    # ==========================================
                    # ★ 終極防禦：429 頻率限制自動重試機制
                    # ==========================================
                    max_retries = 3
                    for attempt in range(max_retries):
                        try:
                            # 使用最省額度的 lite 版，並掛載搜尋工具
                            response = client.models.generate_content(
                                model='gemini-2.5-flash-lite',
                                contents=prompt,
                                config={"tools": [{"google_search": {}}]}
                            )

                            json_str = re.search(r'\{.*\}', response.text, re.DOTALL).group()
                            ai_answers = json.loads(json_str)
                            ai_success = True
                            print(f"✨ AI 雙題庫爬文解答完成！開始填寫！")
                            break  # 成功拿到答案，打破重試迴圈

                        except Exception as e:
                            error_msg = str(e)
                            if "429" in error_msg or "RESOURCE_EXHAUSTED" in error_msg:
                                wait_time = 35
                                print(f"⚠️ 觸發 API 頻率限制，乖乖等待 {wait_time} 秒後進行第 {attempt + 1} 次重試...")
                                time.sleep(wait_time)
                            else:
                                print(f"⚠️ API 發生其他異常，放棄重試退回盲猜。錯誤: {error_msg}")
                                break

                except Exception as e:
                    print(f"⚠️ 題庫解析發生錯誤，退回盲猜。錯誤: {e}")

            for i, q_list in enumerate(question_lists):
                inputs = q_list.find_elements(By.XPATH, ".//input")
                if not inputs: continue
                if ai_success and str(i) in ai_answers:
                    for idx in ai_answers[str(i)]:
                        if idx < len(inputs): driver.execute_script("arguments[0].click();", inputs[idx])
                else:
                    if inputs[0].get_attribute("type") == "checkbox":
                        for opt in random.sample(inputs, random.randint(1, len(inputs))): driver.execute_script("arguments[0].click();", opt)
                    else: driver.execute_script("arguments[0].click();", random.choice(inputs))

            print("👉 所有題目填答完畢，準備最終交卷！")
            time.sleep(5)

            try:
                submit_btn = driver.find_element(By.XPATH, "//input[contains(@value, '結束測驗') or contains(@value, '送出') or contains(@value, '繳交')]")
                driver.execute_script("arguments[0].click();", submit_btn)
            except Exception as e:
                print("❌ 點擊交卷按鈕失敗", e)

            print("⏳ 考卷已最終送出！系統正在結算成績...")
            time.sleep(20)
            print("🎉 結算程序完成！準備返回首頁更新狀態！")
            return True

        return False
    except Exception as e:
        print(f"⚠️ 測驗模組異常: {e}")
        return False

# ==========================================
# ★ 自動問卷引擎 (全階層防彈出版)
# ==========================================
def execute_auto_questionnaire(driver):
    print("\n📝 啟動自動問卷模組...")
    try:
        if switch_to_frame_by_name(driver, "s_main"):
            driver.execute_script("window.location.href = '/learn/questionnaire/questionnaire_list.php';")
        time.sleep(3)
        if switch_to_frame_by_name(driver, "s_main"):

            # 注射防彈出疫苗
            anti_popup_js = """
            function blockPopup(w) {
                try {
                    if (!w.hasInterceptedOpen) {
                        w.originalOpen = w.open;
                        w.open = function(url, name, features) { return w.originalOpen(url, name, ""); };
                        w.hasInterceptedOpen = true;
                    }
                } catch(e) {}
            }
            blockPopup(window); blockPopup(window.parent); blockPopup(window.top);
            window.alert=function(){return true;}; window.confirm=function(){return true;};
            """
            driver.execute_script(anti_popup_js)

            try:
                start_btn = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'process-btn') and contains(., '填寫問卷')]")))
                old_windows = driver.window_handles
                driver.execute_script("arguments[0].click();", start_btn)
            except: return False
            time.sleep(4)

            for window in driver.window_handles:
                if window not in old_windows: driver.switch_to.window(window); break

            driver.execute_script(anti_popup_js) # 進入新分頁再次注射

            time.sleep(4)
            print("🤖 問卷已展開！開始閃電填寫 (全勾選第一個選項)...")
            for q_list in driver.find_elements(By.XPATH, "//ol"):
                try: driver.execute_script("arguments[0].click();", q_list.find_element(By.XPATH, ".//li[1]//input"))
                except: pass

            try:
                submit_btn = driver.find_element(By.XPATH, "//input[contains(@value, '確定繳交') or contains(@value, '送出')]")
                driver.execute_script("arguments[0].click();", submit_btn)
            except: pass

            print("⏳ 問卷已送出！等待系統記錄...")
            time.sleep(8)
            print("🎉 問卷填寫完成！")
            return True
        return False
    except: return False

# ==========================================
# 掛機主程式 (防頻繁登出 + 5分鐘冷卻先上鎖 + 雷達精準辨識)
# ==========================================
def run_bot(login_method, account, password, pincode, api_key, auto_quiz_enabled, auto_survey_enabled):
    global bot_driver, is_hidden, chrome_hwnd
    try:
        skipped_courses = set()
        finished_quiz_timestamp = {}
        finished_survey_timestamp = {}
        all_finished = False

        while not all_finished:
            if stop_event.is_set(): break

            options = webdriver.ChromeOptions()

            # 1. 必備的背景掛機參數
            options.add_argument("--autoplay-policy=no-user-gesture-required")
            options.add_argument("--disable-background-timer-throttling")
            options.add_argument("--disable-backgrounding-occluded-windows")
            options.add_argument("--disable-renderer-backgrounding")

            # ==========================================
            # ★ 解決影片亂跑的 3 大關鍵參數
            # ==========================================
            # 2. 強制視窗最大化與固定解析度，防止播放器 RWD 排版錯亂一直縮放
            options.add_argument("--start-maximized")
            options.add_argument("--window-size=1920,1080")

            # 3. 強制靜音，防止瀏覽器因為安全機制阻擋自動播放而導致播放器當機
            options.add_argument("--mute-audio")

            # 4. 加上這行，騙過網頁讓它以為這是一般真人點開的瀏覽器，而不是機器人
            options.add_argument("--disable-blink-features=AutomationControlled")

            # ⚠️ 檢查：確保這裡【沒有】 --disable-gpu 或 --disable-accelerated-video-decode！

            # ==========================================
            # ★ 終極核彈級解鎖：強制癱瘓 Chrome 安全防護
            # ==========================================
            options.add_argument("--disable-web-security") # 關閉網頁安全機制 (終極殺招)
            options.add_argument("--allow-running-insecure-content") # 允許不安全內容
            options.add_argument("--allow-insecure-localhost") # 允許讀卡機本地端連線
            options.add_argument("--ignore-certificate-errors") # 忽略憑證錯誤
            options.add_argument("--unsafely-treat-insecure-origin-as-secure=https://ecpa.dgpa.gov.tw")

            # 針對最新版 Chrome 的 PNA 攔截進行多重打擊
            options.add_argument("--disable-features=PrivateNetworkAccessPermissionPrompt,PrivateNetworkAccessRespectPreflightResults,BlockInsecurePrivateNetworkRequests,IsolateOrigins,site-per-process")

            # 預設允許所有存取
            prefs = {
                "profile.default_content_setting_values.notifications": 1,
                "profile.default_content_setting_values.private_network_client_server": 1,
                "profile.default_content_setting_values.popups": 1,
                "credentials_enable_service": False,
                "profile.password_manager_enabled": False
            }
            options.add_experimental_option("prefs", prefs)

            driver = webdriver.Chrome(options=options)
            bot_driver = driver
            chrome_hwnd = None

            if is_hidden:
                is_hidden = False
                toggle_browser_visibility()

            wait = WebDriverWait(driver, 15)

            try:
                main_window_handle = login_ecpa(driver, login_method, account, password, pincode)

                while True:
                    if stop_event.is_set(): raise Exception("🛑 收到使用者停止指令！")

                    for window in driver.window_handles:
                        if window != main_window_handle:
                            try:
                                driver.switch_to.window(window)
                                driver.close()
                            except: pass
                    driver.switch_to.window(main_window_handle)

                    print("\n" + "="*40)
                    print("進入課程列表，強制刷新快取掃描中...")

                    timestamp_now = int(time.time())
                    driver.get(f"https://elearn.hrd.gov.tw/mooc/user/learn_dashboard.php?tab=1&_t={timestamp_now}")
                    smart_sleep(5)

                    course_found = False
                    selected_course_link = None
                    selected_course_title = ""
                    cert_hours = 0.0

                    while not course_found:
                        if stop_event.is_set(): raise Exception("🛑 收到使用者停止指令！")

                        if not driver.find_elements(By.ID, "listtype_list_content"):
                            raise Exception("找不到課程列表，可能已斷線。")

                        course_list = driver.find_elements(By.CSS_SELECTOR, "div.course-list-block div.course-list-inner")
                        for course in course_list:
                            try:
                                title_elem = course.find_element(By.XPATH, ".//div[contains(@class, 'cb-info-name1')]")
                                title = title_elem.get_attribute("innerText").strip() if title_elem.get_attribute("innerText") else "未知課程"
                            except: title = "未知課程"

                            if title in skipped_courses: continue

                            course_html = course.get_attribute("innerHTML") or ""

                            if "組裝課程" in course_html or "-- 小時" in course_html:
                                print(f"⏩ 掃描【{title[:10]}...】-> 🛑 偵測為組裝課程或無效時數，跳過。")
                                skipped_courses.add(title)
                                continue

                            is_reading_done = False
                            is_exam_done = False
                            has_exam = "測驗" in course_html
                            is_questionnaire_done = False
                            has_questionnaire = "問卷" in course_html

                            # ==========================================
                            # ★ 精準狀態解析：嚴格檢查單一標籤，避免混淆
                            # ==========================================
                            step_blocks = course_html.split('<div class="step')
                            for block in step_blocks[1:]:
                                short_block = block[:150]
                                if '<div class="label">閱讀' in short_block:
                                    if "done" in short_block or "✔" in short_block: is_reading_done = True
                                if '<div class="label">測驗' in short_block:
                                    if "done" in short_block or "✔" in short_block: is_exam_done = True
                                if '<div class="label">問卷' in short_block:
                                    if "done" in short_block or "✔" in short_block: is_questionnaire_done = True

                            # ==========================================
                            # ★ 最終成績防線：抽取純文字 (innerText) 避開 &nbsp; 陷阱
                            # ==========================================
                            try:
                                pass_infos = course.find_elements(By.CSS_SELECTOR, ".course-list-block-info-pass > div")
                                for info in pass_infos:
                                    info_text = info.get_attribute("innerText") or ""
                                    if "測驗分數" in info_text:
                                        if " 0 分" in info_text or "-- 分" in info_text: is_exam_done = False
                                    elif "問卷狀態" in info_text:
                                        if "未填寫" in info_text or "---" in info_text: is_questionnaire_done = False
                            except: pass

                            now = time.time()
                            if title in finished_quiz_timestamp:
                                if now - finished_quiz_timestamp[title] < 300: is_exam_done = True

                            if title in finished_survey_timestamp:
                                if now - finished_survey_timestamp[title] < 300: is_questionnaire_done = True

                            status_str = f"閱讀: {'✅' if is_reading_done else '❌'} | "
                            status_str += f"測驗: {'無' if not has_exam else ('✅' if is_exam_done else '❌')} | "
                            status_str += f"問卷: {'無' if not has_questionnaire else ('✅' if is_questionnaire_done else '❌')}"
                            print(f"🔎 掃描【{title[:10]}...】-> {status_str}")

                            exam_ok = (not has_exam) or is_exam_done
                            question_ok = (not has_questionnaire) or is_questionnaire_done

                            if is_reading_done and exam_ok and question_ok:
                                continue

                            try:
                                hours_div = course.find_element(By.XPATH, ".//div[contains(text(), '認證時數')]/following-sibling::div")
                                hours_match = re.search(r'[\d\.]+', hours_div.text)
                                temp_cert_hours = float(hours_match.group()) if hours_match else 0.0
                            except: temp_cert_hours = 0.0

                            selected_course_link = course.find_element(By.XPATH, ".//a[1]")
                            selected_course_title = title
                            cert_hours = temp_cert_hours
                            course_found = True
                            break

                        if course_found:
                            break

                        next_page_btns = driver.find_elements(By.XPATH, "//div[@id='pageToolbar']//a[@title='下一頁']")
                        if next_page_btns and "disabled" not in next_page_btns[0].get_attribute("class"):
                            print("本頁處理完畢，翻下一頁...")
                            driver.execute_script("arguments[0].click();", next_page_btns[0])
                            smart_sleep(3)
                        else:
                            print("\n🎉 所有課程皆已完成或進入冷卻中！")
                            all_finished = True
                            break

                    if all_finished:
                        break

                    print(f"\n✅ 發現需動作：【{selected_course_title[:15]}...】")
                    driver.execute_script("arguments[0].click();", selected_course_link)
                    smart_sleep(2)

                    try:
                        read_time_div = wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'majorstatus')]/div[contains(text(), '閱讀時數')]")))
                        read_tm = read_time_div.get_attribute("innerText")
                        match = re.search(r'(\d+):(\d+):(\d+)', read_tm)
                        h, m, s = map(int, match.groups()) if match else (0,0,0)
                        initial_official_seconds = h * 3600 + m * 60 + s
                    except: initial_official_seconds = 0

                    target_seconds = cert_hours * 3600 / 2

                    go_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//*[contains(text(), '上課去')]")))
                    driver.execute_script("arguments[0].click();", go_btn)
                    smart_sleep(3)

                    try: WebDriverWait(driver, 2).until(EC.alert_is_present()).accept()
                    except: pass

                    for window in driver.window_handles:
                        if window != main_window_handle:
                            driver.switch_to.window(window)
                            break
                    smart_sleep(5)

                    session_accumulated_seconds = 0
                    lesson_counter = 0

                    while initial_official_seconds + session_accumulated_seconds < target_seconds:
                        if stop_event.is_set(): raise Exception("🛑 收到使用者停止指令！")
                        if not find_menu_frame(driver): raise Exception("找不到選單，可能已斷線...")

                        ignore_list = ["環境檢測", "課程首頁", "新手上路", "課程資訊", "課程總測", "課程結語", "單元簡介", "小結", "小試身手"]
                        all_lessons = driver.find_elements(By.CLASS_NAME, "cssAnchor1")
                        valid_lessons = []

                        for lesson in all_lessons:
                            l_title = lesson.get_attribute("title")
                            if l_title and not any(ignore in l_title for ignore in ignore_list):
                                valid_lessons.append(lesson)

                        if not valid_lessons: break

                        unfinished = []
                        for lesson in valid_lessons:
                            try:
                                icon_div = lesson.find_element(By.XPATH, "./ancestor::span//div[contains(@class, 'icon-node')]")
                                if "node-finish" not in (icon_div.get_attribute("class") or ""):
                                    unfinished.append(lesson)
                            except: unfinished.append(lesson)

                        chosen_lesson = random.choice(unfinished) if unfinished else random.choice(valid_lessons)
                        lesson_title = chosen_lesson.get_attribute("title").strip()

                        print(f"🎯 選取單元：【{lesson_title[:15]}...】")
                        driver.execute_script("arguments[0].click();", chosen_lesson)
                        smart_sleep(5)

                        wait_seconds = random.randint(180, 500)
                        print(f"⏳ 掛機中，預計停留 {wait_seconds//60} 分鐘... (目前預估: {(initial_official_seconds+session_accumulated_seconds)/3600:.2f}H / 目標: {target_seconds/3600:.2f}H)")

                        for remaining in range(wait_seconds, 0, -60):
                            anti_idle_and_play(driver)
                            smart_sleep(min(60, remaining))

                        session_accumulated_seconds += wait_seconds
                        lesson_counter += 1

                        if lesson_counter >= 5:
                            print("🔄 已累積觀看 5 個單元，切換回背景刷新主畫面防斷線...")
                            driver.switch_to.window(main_window_handle)
                            driver.refresh()
                            smart_sleep(3)
                            if not driver.find_elements(By.ID, "listtype_list_content"):
                                raise Exception("背景刷新時發現斷線...")
                            for window in driver.window_handles:
                                if window != main_window_handle:
                                    driver.switch_to.window(window)
                                    break
                            lesson_counter = 0

                    print("🚀 時數已達標，檢查測驗與問卷...")

                    now = time.time()
                    if has_exam and not is_exam_done:
                        if selected_course_title not in finished_quiz_timestamp or (now - finished_quiz_timestamp[selected_course_title] > 600):
                            if auto_quiz_enabled:
                                print("🔒 [防呆機制] 進入測驗前，優先寫入 10 分鐘冷卻鎖定！")
                                finished_quiz_timestamp[selected_course_title] = time.time()
                                execute_auto_quiz(driver, api_key, selected_course_title)
                            else:
                                print("🛑 自動測驗未啟用，已將此課程加入黑名單跳過。")
                                skipped_courses.add(selected_course_title)
                        else:
                            print("❄️ 測驗剛執行過，處於 5 分鐘冷卻期中，跳過避免重複作答。")

                    now = time.time()
                    if has_questionnaire and not is_questionnaire_done:
                        if selected_course_title not in finished_survey_timestamp or (now - finished_survey_timestamp[selected_course_title] > 600):
                            if auto_survey_enabled:
                                print("🔒 [防呆機制] 進入問卷前，優先寫入 10 分鐘冷卻鎖定！")
                                finished_survey_timestamp[selected_course_title] = time.time()
                                execute_auto_questionnaire(driver)
                            else:
                                print("🛑 自動問卷未啟用，已將此課程加入黑名單跳過。")
                                skipped_courses.add(selected_course_title)
                        else:
                            print("❄️ 問卷剛執行過，處於 10 分鐘冷卻期中，跳過避免重複作答。")

                    print("🔄 本次任務結束，準備返回首頁進行下一輪掃描...")

            except Exception as e:
                if "停止指令" in str(e): break
                print(f"🔄 發生異常，重啟中：{e}")
                try: driver.quit()
                except: pass
                smart_sleep(5)
                continue

        if all_finished: print("\n🏆 任務圓滿達成！所有課程皆已處理完畢。")
    finally:
        btn_start.config(state=tk.NORMAL)
        btn_pause.config(state=tk.DISABLED, text="⏸️ 暫停")
        btn_stop.config(state=tk.DISABLED)

# ==========================================
# UI 介面設計與事件綁定
# ==========================================
def start_bot_thread():
    method = var_login_method.get()
    acc = entry_account.get().strip()
    pwd = entry_password.get().strip()
    pin = entry_pin.get().strip()
    api = entry_apikey.get().strip()
    quiz_on = var_auto_quiz.get()
    sv_on = var_auto_survey.get()

    if method in ["account", "egov"] and (not acc or not pwd):
        messagebox.showwarning("錯誤", "請輸入帳號與密碼！")
        return
    if method in ["moica", "egov_moica"] and not pin:
        messagebox.showwarning("錯誤", "請輸入憑證 PIN 碼！")
        return

# ★ 加入 Try-Except 保護機制，就算沒權限存檔也不會崩潰
    if var_remember.get():
        try:
            with open('pass.txt', 'w', encoding='utf-8') as file:
                json.dump({
                    "method": method, "acc": acc, "pwd": pwd,
                    "pin": pin, "api": api, "quiz": quiz_on, "sv": sv_on
                }, file)
        except PermissionError:
            print("⚠️ 警告：無法儲存設定 (pass.txt 被佔用或無權限)，但不影響本次掛機！")
        except Exception as e:
            print(f"⚠️ 警告：存檔時發生未知的錯誤 {e}，但不影響本次掛機！")
    else:
        if os.path.exists('pass.txt'):
            try: os.remove('pass.txt')
            except: pass

    stop_event.clear()
    pause_event.clear()

    btn_start.config(state=tk.DISABLED)
    btn_pause.config(state=tk.NORMAL, text="⏸️ 暫停")
    btn_stop.config(state=tk.NORMAL)

    threading.Thread(target=run_bot, args=(method, acc, pwd, pin, api, quiz_on, sv_on), daemon=True).start()

def toggle_pause():
    if pause_event.is_set():
        pause_event.clear()
        btn_pause.config(text="⏸️ 暫停")
        print("\n▶️ [系統] 解除暫停，繼續執行...")
    else:
        pause_event.set()
        btn_pause.config(text="▶️ 繼續")
        print("\n⏸️ [系統] 進入暫停模式... (最晚 0.5 秒內生效)")

def stop_bot():
    if messagebox.askyesno("確認", "確定要停止掛機嗎？\n(將會關閉目前的瀏覽器)"):
        stop_event.set()
        pause_event.clear()
        print("\n🛑 [系統] 正在準備停止，請稍候...")
        btn_pause.config(state=tk.DISABLED)
        btn_stop.config(state=tk.DISABLED)

def toggle_browser_visibility():
    global bot_driver, is_hidden, chrome_hwnd
    if bot_driver:
        try:
            if not is_hidden:
                if not chrome_hwnd:
                    original_title = bot_driver.title
                    bot_driver.execute_script("document.title = 'GHOST_MODE_CHROME';")
                    time.sleep(0.5)
                    def foreach_window(hwnd, lParam):
                        global chrome_hwnd
                        if ctypes.windll.user32.IsWindowVisible(hwnd):
                            length = ctypes.windll.user32.GetWindowTextLengthW(hwnd)
                            buff = ctypes.create_unicode_buffer(length + 1)
                            ctypes.windll.user32.GetWindowTextW(hwnd, buff, length + 1)
                            if 'GHOST_MODE_CHROME' in buff.value:
                                chrome_hwnd = hwnd
                                return False
                        return True
                    ctypes.windll.user32.EnumWindows(ctypes.WINFUNCTYPE(ctypes.c_bool, ctypes.POINTER(ctypes.c_int), ctypes.POINTER(ctypes.c_int))(foreach_window), 0)
                    try: bot_driver.execute_script(f"document.title = '{original_title}';")
                    except: pass

                if chrome_hwnd:
                    ctypes.windll.user32.ShowWindow(chrome_hwnd, 0)
                    btn_toggle_browser.config(text="👁️ 顯示視窗")
                    is_hidden = True
                    print("\n👻 瀏覽器已徹底隱藏！")
                else: print("\n❌ 找不到瀏覽器視窗，請稍後再試。")
            else:
                if chrome_hwnd:
                    ctypes.windll.user32.ShowWindow(chrome_hwnd, 5)
                    btn_toggle_browser.config(text="🙈 隱藏視窗")
                    is_hidden = False
                    print("\n👁️ 瀏覽器已切換至前景！")
        except Exception as e: pass
    else:
        is_hidden = not is_hidden
        btn_toggle_browser.config(text="👁️ 顯示視窗" if is_hidden else "🙈 隱藏視窗")

def on_login_method_change():
    method = var_login_method.get()
    if method in ["account", "egov"]:
        entry_account.config(state=tk.NORMAL)
        entry_password.config(state=tk.NORMAL)
        entry_pin.config(state=tk.DISABLED)
    else:
        entry_account.config(state=tk.DISABLED)
        entry_password.config(state=tk.DISABLED)
        entry_pin.config(state=tk.NORMAL)

# ==========================================
# 建立極簡主視窗
# ==========================================
root = tk.Tk()
root.title("e等公務園 自動小幫手")
root.geometry("540x700")
root.configure(padx=15, pady=10)
root.resizable(False, False)

font_title = ("微軟正黑體", 13, "bold")
font_label = ("微軟正黑體", 10)
font_log = ("Consolas", 9)

lbl_title = tk.Label(root, text="💻 e等公務園 幽靈掛機版 (豪華旗艦版)", font=font_title, fg="#128db0")
lbl_title.pack(pady=(0, 10))

frame_inputs = tk.Frame(root)
frame_inputs.pack(fill=tk.X)

var_login_method = tk.StringVar(value="account")
frame_radio = tk.Frame(frame_inputs)
frame_radio.grid(row=0, column=0, columnspan=2, sticky="w", pady=(0, 10))

tk.Radiobutton(frame_radio, text="eCPA帳密", variable=var_login_method, value="account", font=font_label, command=on_login_method_change).pack(side=tk.LEFT, padx=(0, 4))
tk.Radiobutton(frame_radio, text="eCPA自然人憑證", variable=var_login_method, value="moica", font=font_label, command=on_login_method_change).pack(side=tk.LEFT, padx=4)
tk.Radiobutton(frame_radio, text="E政府帳密", variable=var_login_method, value="egov", font=font_label, command=on_login_method_change).pack(side=tk.LEFT, padx=4)
tk.Radiobutton(frame_radio, text="E政府自然人憑證", variable=var_login_method, value="egov_moica", font=font_label, command=on_login_method_change).pack(side=tk.LEFT, padx=4)

lbl_account = tk.Label(frame_inputs, text="帳號：", font=font_label)
lbl_account.grid(row=2, column=0, sticky="e", pady=3)
entry_account = tk.Entry(frame_inputs, font=font_label, width=32)
entry_account.grid(row=2, column=1, pady=3, padx=2, sticky="w")

lbl_password = tk.Label(frame_inputs, text="密碼：", font=font_label)
lbl_password.grid(row=3, column=0, sticky="e", pady=3)
entry_password = tk.Entry(frame_inputs, font=font_label, show="*", width=32)
entry_password.grid(row=3, column=1, pady=3, padx=2, sticky="w")

lbl_pin = tk.Label(frame_inputs, text="憑證 PIN 碼：", font=font_label)
lbl_pin.grid(row=4, column=0, sticky="e", pady=3)
entry_pin = tk.Entry(frame_inputs, font=font_label, show="*", width=32, state=tk.DISABLED)
entry_pin.grid(row=4, column=1, pady=3, padx=2, sticky="w")

lbl_apikey = tk.Label(frame_inputs, text="Gemini API：", font=font_label)
lbl_apikey.grid(row=5, column=0, sticky="e", pady=3)
entry_apikey = tk.Entry(frame_inputs, font=font_label, width=32)
entry_apikey.grid(row=5, column=1, pady=3, padx=2, sticky="w")

# 自動測驗與問卷選項
frame_opts = tk.Frame(frame_inputs)
frame_opts.grid(row=6, column=0, columnspan=2, pady=8)

var_auto_quiz = tk.BooleanVar(value=True)
chk_auto_quiz = tk.Checkbutton(frame_opts, text="自動測驗", variable=var_auto_quiz, font=font_label, fg="#d11f11")
chk_auto_quiz.pack(side=tk.LEFT, padx=10)

var_auto_survey = tk.BooleanVar(value=True)
chk_auto_survey = tk.Checkbutton(frame_opts, text="自動問卷", variable=var_auto_survey, font=font_label, fg="#d11f11")
chk_auto_survey.pack(side=tk.LEFT, padx=10)

var_remember = tk.BooleanVar(value=True)
chk_remember = tk.Checkbutton(frame_opts, text="記住所有設定", variable=var_remember, font=font_label)
chk_remember.pack(side=tk.LEFT, padx=10)

if os.path.exists('pass.txt'):
    try:
        with open('pass.txt', 'r', encoding='utf-8') as file:
            data = json.load(file)
            var_login_method.set(data.get("method", "account"))
            entry_account.insert(0, data.get("acc", ""))
            entry_password.insert(0, data.get("pwd", ""))
            entry_pin.insert(0, data.get("pin", ""))
            entry_apikey.insert(0, data.get("api", ""))
            var_auto_quiz.set(data.get("quiz", True))
            var_auto_survey.set(data.get("sv", True))
            on_login_method_change()
    except: pass

frame_buttons = tk.Frame(root)
frame_buttons.pack(pady=10)

style = ttk.Style()
style.configure("TButton", font=("微軟正黑體", 9, "bold"))

btn_start = ttk.Button(frame_buttons, text="▶ 開始掛機", command=start_bot_thread, width=11)
btn_start.grid(row=0, column=0, padx=4)

btn_pause = ttk.Button(frame_buttons, text="⏸️ 暫停", command=toggle_pause, width=9, state=tk.DISABLED)
btn_pause.grid(row=0, column=1, padx=4)

btn_stop = ttk.Button(frame_buttons, text="⏹️ 停止", command=stop_bot, width=9, state=tk.DISABLED)
btn_stop.grid(row=0, column=2, padx=4)

btn_toggle_browser = ttk.Button(frame_buttons, text="🙈 隱藏視窗", command=toggle_browser_visibility, width=12)
btn_toggle_browser.grid(row=0, column=3, padx=4)

txt_log = scrolledtext.ScrolledText(root, wrap=tk.WORD, font=font_log, bg="#1e1e1e", fg="#00ff00", height=15)
txt_log.pack(fill=tk.BOTH, expand=True)

sys.stdout = PrintRedirector(txt_log)
sys.stderr = PrintRedirector(txt_log)

print("✅ 系統就緒。")
print("💡 提示：本程式使用Google Chrome進行掛機，請先確認是否安裝！")
print("💡 提示：使用自然人憑證前，請確保讀卡機已接上且卡片已插入！")
print("💡 提示：輸入 Gemini API Key 並勾選測驗/問卷，即可啟動全自動通關！")
print("💡 提示：API Key可免費申請 https://aistudio.google.com/app/api-keys")

lbl_author = tk.Label(root, text="製作人：程門立雪 & Gemini by 2026/04", font=("微軟正黑體", 9), fg="#888888")
lbl_author.pack(side=tk.BOTTOM, pady=(5, 0))

root.mainloop()
```
